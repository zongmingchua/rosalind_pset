```python
from itertools import permutations
```

```python
def perms(n):
    
    x = []
    
    #x becomes a list of n elements from 1 to n
    for i in range(1, n+1):
        x.append(i)

    perms = []

    #make a list of all permutations
    for i in permutations(x):
        perms.append(i)
        
    print (len(perms))
    
    for i in perms:
        print(*i)
    
    
```

```python
perms(5)
```
