```python
#second attempt at code
#takes in total num of nodes and list of edges and returns number of new edges needed to form tree
#note edges used here are strings and function is agnostic to int or str edges!
def tree(n, edges):
    
    #list of groups- new edges is group num minus one
    groups = []
    
    #iterate over each edge,
    for edge in edges:
           
        #let the two nodes of each edge be a and b
        # a/b represents the index of any group that the node is part of 
        a = None
        b = None
               
        #iterate over all groups
        for i in range(0, len(groups)):
            
            if edge[0] in groups[i]: 
                a = i
                
            elif edge[1] in groups[i]:
                b = i
                
        #only a connects
        if (a is not None) and (b is None):
            groups[a].append(edge[1])
        
        #only b connects
        elif (a is None) and (b is not None):
            groups[b].append(edge[0])
            
        #both a and b connect - merge the groups 
        elif (a is not None) and (b is not None):
            #modify first than remove!
            groups[a] = groups[a]+groups[b]
            groups.pop(b)
        
        #neither a or b connects
        else:
            groups.append(edge)
        
    new_edge_count = len(groups)-1
    
    #connect nodes not mentioned in edge list 
    count = 0
    for group in groups:
        
        for node in group:
            count +=1
            
    #print(n, len(groups),count, groups)
    
    single_nodes_count = n-count
    
    return new_edge_count+single_nodes_count
            
```
