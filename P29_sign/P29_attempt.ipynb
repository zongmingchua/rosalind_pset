```python
import math
from itertools import *
```

```python
#outputs number of possible signed permutations and then all permutations
def sign(n):
    
    #number of permutations possible
    count_perms = math.factorial(n) *  2**n
    
    #generate initial range
    ns = list(range(1,n+1))
    
    #make neg signs
    for i in range(0,len(ns)):
        ns[i] = [ns[i], -ns[i]]
        
    #generate perms 
    perm_ns = permutations(ns)
        
    #makes 1, 0 list to call
    a = list(product([0,1],repeat=n))
    
    perm_list = []
    
    #first iterate over all perms of number range
    for perm in perm_ns:
        
        #then iterate over the + / - toggle list
        for binary in a:
            new_perm = []            
            
            #iterate over each toggle
            for i in range(0, len(binary)):
                toggle = binary[i]
                new_perm.append(perm[i][toggle])
                
            #add to perm_list each new perm           
            perm_list.append(new_perm)
            
    
    return count_perms, perm_list
    
    
```
