```python
def revc(dna_in):
    dna_out = ''
    for i in dna_in:
        if i == 'A':
            dna_out = 'T' + dna_out  
        elif i == 'T':
            dna_out = 'A' + dna_out
        elif i == 'C':
            dna_out = 'G' +  dna_out 
        elif i == 'G':
            dna_out = 'C' + dna_out
        else:
            print ('non-canonical base!')
            break
        #print(i)
    print(dna_out)    
```
