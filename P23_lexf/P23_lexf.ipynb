```python
from itertools import product
#many ways to iterate
#see https://docs.python.org/3/library/itertools.html 
```

```python
with open('rosalind_lexf.txt') as txt:
    content = txt.read().splitlines()
    check = 1
    
    #obtain both lines
    for line in content:
        if check == 1:
            letters = line.split()
            check +=1
            
        else: 
            perm_no = int(line)
    
```

```python
#sorts letters alphabetically automatically
letters.sort()
```

```python
perms = product(letters, repeat = perm_no)

#print each permutation
for i in perms:
    print(''.join(i)) #joining list elements together 
```
